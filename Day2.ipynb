```python
# Set 집합
# {} 를 사용하여 작성한다. 
# 중복을 허용하지 않는다. 
# 순서가 없다. unordered 리스트나 튜플은 순서가 있기때문에 ordered  인덱싱과 슬라이싱을 통해
# 값을 꺼낼수 있지만 딕셔너리와 셋은 순서가 없어 불가능하다. 
# 중복을 허용하지 않는다는 특성을 활용하여 중복 제거 필터용으로 많이 사용된다. 
# set자료형에 저장된 값을 꺼내려면 리스트나 튜플로 변환하여 사용한다.
# mutable 수정/삭제가 가능하다. 
set = {1,1,2,2,2,3,4}
print(set)

print(set[0])
```

```python
# 집합 계산

set1 = {1,2,3}
set2 = {2,3,4}

print(type(set1))
print(type(set2))

# 양쪽 집합에 모두 공통적으로 포함되는 원소 -> 교집합
print(set1 & set2)  
print(set1.intersection(set2))

# 양쪽 집합의 모든 원소 -> 합집합
print(set1 | set2)
print(set1.union(set2))

# 두개 집합의 차 -> 차집합
print(set1 - set2)
print(set1.difference(set2))  
```

```python
# 집합은 삽입 가능 mutable

set1 = {1,2,3}
print(set1)
set1.add(4)
print(set1)

set1.update([5,6,7,8])
print(set1)

#집합은 삭제가 가능하다.  mutable
#집합은 인덱스 개념이 존재하지 않는다. 따라서 값을 지정하여 직접 삭제 한다. 
set1.remove(4)
print(set1)
```

```python
# 임의의 정수를 입력받아서 그 수가 3의 배수인지를 확인하는
# 프로그램을 작성하시오 % 홀수 짝수 배수

# (입력) 정수를 입력하세요 >>> 14
# (출력) 14는 3의 배수가 아닙니다.

# (입력) 정수를 입력하세요 >>> 15
# (출력) 15는 3의 배수입니다.

n = int(input("정수를 입력하세요"))
r = ""
if n%3 !=0:
  r = "가 아닙니다"
else: 
  r ="입니다"

print(f"{n}은 3의 배수{r}")
```
